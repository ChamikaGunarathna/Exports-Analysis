## Importing Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
exports_data = pd.read_csv('exports_data.csv')
```

```python
exports_data
```

Looking at data we can see that coconuts exports are given by country and also by the amount of nuts.

Therefore we can break this into **3 tables**.

1.  **Tea, Rubber, Garment and Coconuts total exports (US $ mn)**
2.  **Coconut Exports by Country (US $ mn)**
3.  **Coconut production and exports by amount of nuts (mn nuts)**

## Splitting Data set into 3 tables

```python
total_exports = exports_data.iloc[:3,:].reset_index(drop='index')
ccnut_exp_nut = exports_data.iloc[4:6,:].reset_index(drop='index')
ccnt_exp_country = exports_data.iloc[7:,:].reset_index(drop='index')
```

```python
display(total_exports)
display(ccnut_exp_nut)
display(ccnt_exp_country)
```

Getting total coconut exports from **table 3** and adding them to the **table 1**

```python
ccnt_exp_country.loc[15] = ['total_exports' ]+ list(np.zeros(7))
years = ccnt_exp_country.set_index('Product').columns
for column in years:
    ccnt_exp_country.loc[15, column] = ccnt_exp_country[column].sum()
total_exports_ccnts = ccnt_exp_country.iloc[15:,:].set_index('Product')
total_exports_ccnts.rename(index={'total_exports':'Coconuts Exports (US $ mn)'},inplace=True)
total_exports =total_exports.set_index('Product')
total_exports = pd.concat([total_exports, total_exports_ccnts])
total_exports
```

### Analysis - Total Exports

```python
total_exports
```

```python
years = total_exports.columns
years
```

```python
tea = total_exports.loc[total_exports.index[0],:].values
rubber = total_exports.loc[total_exports.index[1],:].values
garments = total_exports.loc[total_exports.index[2],:].values
coconuts = total_exports.loc[total_exports.index[3],:].values
```

```python
bins = np.arange(len(years))
width = 0.2 #1/6

plt.figure(figsize=(10,8))
plt.bar(bins,rubber,color='red',width=width,label='Rubber')
plt.bar(bins+width,tea,color='blue',width=width,label='Tea')
plt.bar(bins+2*width,garments,color='green',width=width,label='Garment')
plt.bar(bins+3*width,coconuts,color='yellow',width=width,label='Coconuts')

plt.xlabel("Year") 
plt.ylabel("Exports (US $ mn)") 
plt.title("Exports in each year")

plt.xticks(bins + 1.5*width,years) 
plt.legend()
```

```python
#Showing the percentages over the years
total_exports_per = total_exports.copy()
for year in years:
    total_exports_per[year] = (total_exports[year]/total_exports[year].sum())*100
total_exports_per
```

```python
#getting average exports
totat_exports_stats= total_exports_per.transpose()
stats = totat_exports_stats.describe()
total_exports_per['Average'] = stats.mean()
total_exports_per = total_exports_per.reset_index()

labels = 'Tea', 'Rubber', 'Garment', 'Coconut'
sizes = total_exports_per['Average']

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels,autopct='%1.1f%%')
```

```python
#getting total sums of the exports
total_exports_sum = total_exports.copy()
```

```python
total_exports_sum
```

```python
total_exp_over_years =[]
for i in range(4):
    total_exp_over_years.append(total_exports_sum.loc[total_exports_sum.index[i],:].values.sum())
total_exp_over_years
```

```python
total_exports_sum['total_over_years'] = total_exp_over_years
total_exports_sum
```

```python
total_exports_sum['avg_per_year'] = total_exports_sum['total_over_years']/7
total_exports_sum
```

```python
labels = 'Tea', 'Rubber', 'Garment', 'Coconut'
sizes = total_exports_sum['avg_per_year']

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels,autopct='%1.1f%%')
```
